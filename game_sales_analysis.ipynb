```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
game_file = "/Users/vandanpatel/Desktop/Data Analysis/vgsales.csv"
# file with software sales

console_file = "/Users/vandanpatel/Desktop/Data Analysis/console.csv"
# file with console sales
```

Both data set found on Kaggle

Sales are in millions

```python
df = pd.read_csv(game_file)
df.head()
```

```python
df2 = pd.read_csv(console_file)
df2.head()
```

Now I want to filter down to the games where global sales are at least 15 million

```python
top_global = df.loc[df['Global_Sales'] >= 15]
```

```python
top_global
```

Now with the top games sold filtered out, I wanted to create a visualization showcasing this data.

```python
labels = top_global["Publisher"].unique()
publisher_count = top_global["Publisher"].value_counts()
publisher_count

#creating pie chart
colors = ["Red", "Green", "Blue"]
explode = (0.1, 0.1, 0.1)
plt.title("Top 3 Game Publishers")
plt.pie(publisher_count, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=90)
plt.axis("equal")
plt.show()
```

Looks like Nintendo has the most amount of global software sales over the competition.
Now I want to see if the same thing can be said for hardware sales.

```python
df2
```

Dropping PC row as it does not count as a traditional home video game console

```python
df2 = df2.drop(labels=0, axis=0)
```

```python
df2
```

```python
labels = df2["Manufacturer"].unique()
manufacturer_count = df2["Manufacturer"].value_counts()
manufacturer_count
plt.title("Total Global Hardware Sales by Manufacturer", loc="left")
plt.rcParams["figure.figsize"] = [12, 9]
plt.rcParams["figure.autolayout"] = True
plt.pie(manufacturer_count, labels=labels, autopct="%1.1f%%", shadow=True, startangle=100)
plt.axis("equal")
plt.show()
```

We can see that Sony has the most amount of hardware sold than any other company.
They have double the sales of the second highest seller, Nintendo.

When we compare the data, it is interesting to see that even though Sony leads in global hardware sales, Nintendo absolutely dominates in software sales.
